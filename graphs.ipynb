```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("df_final.csv")
```

```python
sns.boxplot(df["stress"])
plt.ylabel("%")
plt.title("Water Stress (%)")
```

```python
sns.boxplot(df[df["stress"] < 200]["stress"])
plt.ylabel("%")
plt.title("Water Stress (%)")
```
