```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_final = pd.read_csv("df_final.csv")
```

## Correlations and Multicollinearity

```python
sns.heatmap(df_final.corr(numeric_only=True), vmin = -1, vmax = 1, cmap = 'RdBu')
```

```python
abs(df_final.corr(numeric_only=True)) >= 0.6
```

## Boxplots

```python
for col in df_final[['year',
                     'stress',
                     'agriculture_water',
                     'industry_water',
                     'domestic_water',
                     'water_per_capita',
                     'urban_population',
                     'state_capacity',
                     'cattle',
                     'total_withdrawals',
                     'cereal_yield',
                     'control_of_corruption',
                     'hydro_electricity',
                     'permanent_cropland',
                     'population_density'
                     ]]:
    sns.boxplot(data = df_final, x = col)
    plt.show()
```

## Scatterplot of every explanatory variable with the response variable

```python
for col in df_final[['year',
                     'agriculture_water',
                     'industry_water',
                     'domestic_water',
                     'water_per_capita',
                     'urban_population',
                     'state_capacity',
                     'cattle',
                     'total_withdrawals',
                     'cereal_yield',
                     'control_of_corruption',
                     'hydro_electricity',
                     'permanent_cropland',
                     'population_density'
                     ]]:
    sns.scatterplot(data = df_final, x = col, y = 'stress')
    plt.show()
```
